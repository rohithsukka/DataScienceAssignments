```python
import pandas as pd
import numpy as np

df = pd.read_csv('https://raw.githubusercontent.com/rohithsukka/Datasets/main/playtennis.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df1 = df.drop(['Day'] , axis='columns')
df1.head()
```

```python
print(df1['Outlook'].unique())
print(df1['Temperature'].unique())
print(df1['Humidity'].unique())
print(df1['Wind'].unique())
print(df1['PlayTennis'].unique())
```

```python
df2 = df1.copy()
```

```python
columns = ['Outlook','Temperature', 'Humidity', 'Wind', 'PlayTennis']

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for col in columns:
  df2[col] = le.fit_transform(df2[col])
```
